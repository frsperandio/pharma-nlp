```python
# notebooks/exploracao_anvisa.ipynb

import pandas as pd

# Caminho para o CSV (ajuste se necessário)
csv_path = "../data/inputs/DADOS_ABERTOS_MEDICAMENTOS.csv"

# Leitura da base
df = pd.read_csv(csv_path, sep=';', encoding='ISO-8859-1')

# Visualiza as primeiras linhas
df.head()
```

```python
# Extrair nomes únicos de produtos, em caixa alta e sem espaços extras
nomes_medicamentos = (
    df["NOME_PRODUTO"]
    .dropna()
    .str.upper()
    .str.strip()
    .drop_duplicates()
    .sort_values()
    .tolist()
)

# Ver alguns exemplos
nomes_medicamentos[:50]
```

```python
output_path = "../data/lookup/nomes_medicamentos_anvisa.txt"

with open(output_path, "w", encoding="utf-8") as f:
    for nome in nomes_medicamentos:
        f.write(f"{nome}\n")
```
